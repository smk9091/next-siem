```python
from pyathena import connect

# Athena 연결 설정
conn = connect(
    s3_staging_dir='s3://siem-jupyter/',  # 쿼리 결과가 저장될 S3 경로
    region_name='us-east-1',  # 예: us-east-1, ap-northeast-2 등
    schema_name='siem_real',
)
```

```python
cursor = conn.cursor()
import pandas as pd
query = """
SELECT * FROM gd_logs 
LIMIT 5
"""
cursor.execute(query)
# DataFrame으로 변환
df = pd.DataFrame(cursor.fetchall())

# 결과 출력
display(df)
```
